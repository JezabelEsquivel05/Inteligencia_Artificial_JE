```python
!pip install tensorflowjs
!pip install tensorflow==2.16.0rc0z

import os
os.environ["TF_USE_LEGACY_KERAS"] = "1"

!mkdir predecirCo22

import tensorflow as tf
import numpy as np




# Valores de CO correspondientes a cada hora del día (0-23)
co_values = np.array([
    0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0,
    4.5, 5.0, 5.5, 6.0, 6.5, 7.0, 7.5, 8.0,
    8.5, 9.0, 9.5, 10.0, 10.5, 11.0, 11.5, 12.0
], dtype=float)

# Usamos un rango de 0 a 23 como entrada (horas del día)
entrada = np.array(range(24), dtype=float)

# Definir modelo, con funciones de Keras
modelo = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=[1]),  # Capa de entrada
    tf.keras.layers.Dense(units=5, activation='relu'),  # Capa oculta con activación ReLU
    tf.keras.layers.Dense(units=1)  # Capa de salida
])

modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)

print("Entrenando...")
historial = modelo.fit(entrada, co_values, epochs=300, verbose=False)
print("Modelo entrenado!")

# Guardar el modelo en formato HDF5 en la carpeta 'predecirCO'
modelo.save('co_values.h5')

# Convertir el modelo a TensorFlow.js
!tensorflowjs_converter --input_format=keras co_values.h5  predecirCo22

# Realizar una predicción
hora_prueba = 10.0  # Cambia este valor de entrada según la hora que deseas predecir
resultado = modelo.predict(np.array([hora_prueba]))
print("Predicción resultado es :" + str(resultado[0][0]) + " µg/m³ de CO!")
```
