```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Definir la clase Perceptron
class Perceptron(object):
    def __init__(self, eta=0.01, n_iter=50, random_state=1):  # Cambié _init_ por __init__
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, x, y):
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + x.shape[1])  # Inicialización de los pesos
        self.errors_ = []  # Almacenar el número de errores en cada época

        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(x, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi  # Actualizar los pesos
                self.w_[0] += update  # Actualizar el sesgo
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, x):
        """Calcular el input neto"""
        return np.dot(x, self.w_[1:]) + self.w_[0]

    def predict(self, x):
        """Devolver la clase predicha"""
        return np.where(self.net_input(x) >= 0.0, 1, -1)
v1 = np.array([1, 2, 3])
v2 = 0.5 * v1

np.arccos(v1.dot(v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data', header=None)
df.tail()

```

```python
# Etiquetas y características
y = df.iloc[0:100, 4].values
y = np.where(y == 'Iris-setosa', -1, 1)  # Convertir las etiquetas a -1 y 1


x = df.iloc[0:100, [0, 2]].values

# Graficar los datos
plt.scatter(x[:50, 0], x[:50, 1], color='red', marker='o', label='setosa')
plt.scatter(x[50:100, 0], x[50:100, 1], color='blue', marker='x', label='versicolor')
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')
plt.show()

# Entrenar el modelo de perceptrón
ppn = Perceptron(eta=0.1, n_iter=10)
ppn.fit(x, y)

# Graficar el número de errores por época
plt.plot(range(1, len(ppn.errors_) + 1), ppn.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of updates')
plt.show()
```
